```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('izayoisakuya-monokuro.jpg')
plt.imshow(img)
```

```python
kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(img,-1,kernel)
plt.imshow(dst)
```

```python
dst = cv2.GaussianBlur(img,(3,3),-1)
plt.imshow(dst)
```

```python
img = cv2.imread('izayoisakuya-color.jpg')
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
dst = cv2.medianBlur(img_rgb,3,-1)
plt.imshow(dst)
```

```python
img = cv2.imread('izayoisakuya-color.jpg')
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
img_blured = cv2.GaussianBlur(img_rgb,(11,11),10)
dst = cv2.addWeighted(img_rgb,2,img_blured,-1,0)
plt.imshow(dst)
```
